```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import sqlite3
```

```python
connection = sqlite3.connect(':memory:')
cursor = connection.cursor()
```

```python
adult=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
adult.head()
```

```python
with connection:
    cursor.execute('''CREATE TABLE IF NOT EXISTS adult (
                            age INTEGER,workclass VARCHAR,fnlwgt INTEGER,education VARCHAR,education_num INTEGER,marital_status VARCHAR,
                            occupation VARCHAR,relationship VARCHAR,race VARCHAR,sex VARCHAR,capital_gain INTEGER,capital_loss INTEGER,
                            hours_per_week INTEGER,native_country VARCHAR,oup VARCHAR)''')
    cursor.executemany('INSERT INTO adult VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)',adult.values.tolist())
                       
```

```python
cursor.execute('SELECT * FROM adult LIMIT 10')
print(cursor.fetchall())
```

```python
cursor.execute("SELECT AVG(hours_per_week) as avg_hours from adult where sex=' Male' and workclass=' Private';")
print(cursor.fetchall())
```

```python
cursor.execute("select count(*) as Education_Freq,education from adult group by education;")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT count(*) as occup_relat_freq,occupation,relationship from adult group by occupation,relationship")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * from adult where lower(marital_status) like ' married%' and lower(workclass) like '%private%' and lower(education) like '%masters%';")
print(cursor.fetchall())
```

```python
cursor.execute("select distinct workclass from adult;")
print(cursor.fetchall())
```

```python
cursor.execute("select avg(age) as avg_age_group,min(age) as min_age_group,max(age) as max_age_group,workclass from adult group by workclass;")
print(cursor.fetchall())
```

```python
cursor.execute("select count(age),native_country from adult group by native_country;")
print(cursor.fetchall())
```

```python
cursor.execute("select (capital_gain-capital_loss) as Net_Capital_Gain from adult;")
print(cursor.fetchall())
```
